```python
import requests
from bs4 import BeautifulSoup
import time 
import re
#參考來源 : https://matthung0807.blogspot.com/2019/08/python-3-requests-ptt.html
#網頁ASCII編碼 : https://www.w3school.com.cn/tags/html_ref_urlencode.html
#HTTP 302 : https://zh.wikipedia.org/wiki/HTTP_302
#Github 登入 : https://github.com/Kr1s77/awesome-python-login-model/blob/master/Github/login.py
#歸於同個Session類，保持會話的功能，避免重複request時某些資料備重複洗掉，FormData、token之類可能每次request都不同
#各大網站登入(爬蟲) : https://iter01.com/179722.html (主要參考)

#step1 : 取得token，token為一開始request至登入頁(從瀏覽器跳轉)時就會放在網頁原始碼之中，每次requests 的token 都不一樣
#同樣也可以看 post_url 中那些是一開始就給的payload
rs = requests.Session()

login_url = 'https://github.com/login'
post_url = 'https://github.com/session' #觀察登入後在network多了一個.session，FormData表我們須提供(post)的資料，取得session 授權後會引導至github
headers = { #加上基本的header
            'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98 Safari/537.36',
            'Referer': 'https://github.com/',
            'Host': 'github.com'
        }

response = rs.get(login_url,headers = headers)      
#print(response.url)
#print(response.text)

if response.status_code != 200: #增設檢查點
    print("not get the token")
else :
    match = re.search(r'name="authenticity_token" value="(.*?)"', response.text) #用regex從網頁原始碼中找
    if not match:
        print("not get the token")
    else : 
        at = match.group(1) #at = authenticity_token，at 為目標 
        print(at)
#補充 : 
#re模塊 : https://blog.csdn.net/SoaringLee_fighting/article/details/78988430
```

```python

#依照不同帳號密碼進行登入，若輸入帳密錯誤response.url會是post_url，反之成功的話會是http://github.com
#
ac = input('Account:')
pd = input('Password:')

#取得payload 要post的東西
payload={  #找到session(在開發工具network(歸類在other)中，request from post_url，判斷方法為在Form Data可以看到帳密&要放在payload中的內容)
     'commit': 'Sign in',
     'utf8': '✓',
     'authenticity_token': at,
     'login': ac,
     'password': pd
}


res = rs.post(post_url,verify = False, data = payload) #至登入頁面做登入的動作
print(res.url) #檢查是否登入
print(res.text)
```
